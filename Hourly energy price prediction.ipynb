```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df_price = pd.read_excel("Day-ahead_prices_201901010000_202409290000_Hour.xlsx")
df_price = df_price.drop(index=range(0, 8))
df_price = df_price.reset_index(drop=True)
new_header = df_price.iloc[0]
df_price = df_price[1:]
df_price.columns = new_header 
df_price.head()
```

```python
df_consum = pd.read_excel("Actual_consumption_201901010000_202409290000_Hour.xlsx")
df_consum = df_consum.drop(index=range(0, 8))
df_consum = df_consum.reset_index(drop=True)
new_header = df_consum.iloc[0]
df_consum = df_consum[1:]
df_consum.columns = new_header
df_consum.info()
```

```python
def convert_to_float(x):
    try:
        return float(x)
    except ValueError:
        return np.nan
    
columns_to_convert = ['Total (grid load) [MWh]', 'Residual load [MWh]', 'Hydro pumped storage [MWh]']
for col in columns_to_convert:
    df_consum[col] = df_consum[col].apply(convert_to_float)
#fill the missing value with the mean of particular numeric columns
df_consum[columns_to_convert] = df_consum[columns_to_convert].apply(lambda col: col.fillna(col.mean()))
#normalise the numeric data
scaler = MinMaxScaler()
df_consum[columns_to_convert] = scaler.fit_transform(df_consum[columns_to_convert])
df_consum.info()
```

```python
df_generate = pd.read_excel("Actual_generation_201901010000_202409290000_Hour.xlsx")
df_generate = df_generate.drop(index=range(0, 8))
df_generate = df_generate.reset_index(drop=True)
new_header = df_generate.iloc[0]
df_generate = df_generate[1:]
df_generate.columns = new_header
```

```python
columns_to_convert = ['Biomass [MWh]','Hydropower [MWh]','Wind offshore [MWh]','Wind onshore [MWh]','Photovoltaics [MWh]','Other renewable [MWh]','Nuclear [MWh]','Lignite [MWh]','Hard coal [MWh]','Fossil gas [MWh]','Hydro pumped storage [MWh]','Other conventional [MWh]']
for col in columns_to_convert:
    df_generate[col] = df_generate[col].apply(convert_to_float)
#fill the missing value with the mean of particular numeric columns
df_generate[columns_to_convert] = df_generate[columns_to_convert].apply(lambda col: col.fillna(col.mean()))
#normalise the numeric data
scaler = MinMaxScaler()
df_generate[columns_to_convert] = scaler.fit_transform(df_generate[columns_to_convert])
df_generate.info()
df_generate.head(5)
```

```python
df_merge = pd.merge(df_consum,df_price[['Start date','Germany/Luxembourg [€/MWh]']],on = 'Start date')
df = pd.merge(df_generate,df_merge[['Start date','Germany/Luxembourg [€/MWh]','Total (grid load) [MWh]','Residual load [MWh]','Hydro pumped storage [MWh]']],on = 'Start date')
df = df.rename(columns={'Total (grid load) [MWh]': 'Total Grid','Germany/Luxembourg [€/MWh]':'Germany Prices','Residual load [MWh]':'Residual Load','Hydro pumped storage [MWh]_x':'Hydro pumped G','Hydro pumped storage [MWh]_y':'Hydro pumped C'})
df['Start date'] = pd.to_datetime(df['Start date'])
df['Date'] = pd.to_datetime(df['Start date'], format='%Y-%m-%d',errors='coerce').dt.date
df['Time'] = pd.to_datetime(df['Start date'],format='%H:%M:%S').dt.time
df["Month"] = pd.to_datetime(df["Start date"]).dt.month
df["Day"] = pd.to_datetime(df["Start date"]).dt.day_name()
df.drop(columns={'End date','Start date'},inplace=True)
df.set_index('Date',inplace=True)
df.index = pd.to_datetime(df.index, errors='coerce')
df['Germany Prices'] = pd.to_numeric(df['Germany Prices'], errors='coerce')
df.head(5)
```

```python
def time_to_float(time_obj):
    return time_obj.hour + time_obj.minute / 60 + time_obj.second / 3600
df['Time'] = df['Time'].apply(time_to_float)
```

# Time reshampling

```python
df.resample(rule='A').min()['Germany Prices'].plot(kind='bar')
```

```python
df.resample(rule='M').max()['Germany Prices'].plot(kind='bar',figsize=(15,6))
```

```python
#simple moving average
df['Price:10 days rolling']=df['Germany Prices'].rolling(30).mean()
df['Price:30 days rolling']=df['Germany Prices'].rolling(50).mean()
df[['Germany Prices','Price:10 days rolling','Price:30 days rolling']].plot(xlim=['2020-01-01','2020-03-30'],ylim=[-100,100],figsize=(12,5))
```

```python
#cummulative moving average
df['Germany Prices'].expanding().mean().plot()
```

```python
#exponential moving average(focus more on recent data)
#smoothing factor = 0.1
df['EMA_0.1']=df['Germany Prices'].ewm(alpha=0.1,adjust=False).mean()
df[['Germany Prices','EMA_0.1']].plot(xlim=['2020-01-01','2021-01-01'],ylim=[-100,100],figsize=(12,8))
```

```python
df['EMA_0.5']=df['Germany Prices'].ewm(alpha=0.5,adjust=False).mean()
df[['Germany Prices','EMA_0.5']].plot(xlim=['2020-01-01','2020-10-01'],ylim=[-100,100],figsize=(12,8))
```

```python
#EWMA
df['EWMA_5']=df['Germany Prices'].ewm(span=5).mean()
df['EWMA_half']=df['Germany Prices'].ewm(halflife=5).mean()
df[['Germany Prices','EWMA_5','EWMA_half']].plot(xlim=['2020-01-01','2020-10-30'],ylim=[-200,200],figsize=(12,8))
```
