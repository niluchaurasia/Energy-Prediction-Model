Model Implementation

```python
df.drop(['Date','Time','Day','Quarter'], axis=1,inplace=True)
```

```python
df.replace('-', 0, inplace=True)
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)], lower_bound, upper_bound

outliers_total_consumption, tc_lower, tc_upper = detect_outliers_iqr(df, 'Total Consumption')
outliers_germany_prices, gp_lower, gp_upper = detect_outliers_iqr(df, 'Germany Prices')
```

```python
scaler = MinMaxScaler()

df[['Total Consumption', 'Germany Prices']] = scaler.fit_transform(df[['Total Consumption', 'Germany Prices']])

df_no_outliers = df[(df['Total Consumption'] >= tc_lower) & (df['Total Consumption'] <= tc_upper) &
                    (df['Germany Prices'] >= gp_lower) & (df['Germany Prices'] <= gp_upper)]

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
sns.boxplot(df['Total Consumption'])
plt.title('Total Consumption (After Outlier Removal)')

plt.subplot(1, 2, 2)
sns.boxplot(df['Germany Prices'])
plt.title('Germany Prices (After Outlier Removal)')

plt.tight_layout()
plt.show()
```

```python
features = ['Total Consumption', 'Residual load', 'Hydro pumped storage']
target = 'Germany Prices'
df.set_index('Start date', inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

scaled_X = scaler_X.fit_transform(df[features])
scaled_y = scaler_y.fit_transform(df[[target]])
```

```python
scaled_y = scaled_y.reshape(-1)
```

```python
def create_lstm_dataset(X, y, time_steps=10):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps), :])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
# Create the dataset with a lookback window of 10 hours
time_steps = 1
X_lstm, y_lstm = create_lstm_dataset(scaled_X, scaled_y, time_steps)
X_lstm, y_lstm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_lstm, y_lstm, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')
```

```python
predictions = model.predict(X_test)
```

```python
predictions = scaler_y.inverse_transform(predictions)
```

```python
print("Predictions:\n", predictions)
```

```python
y_test_actual = scaler_y.inverse_transform(y_test.reshape(-1, 1))
```

```python
mse = mean_squared_error(y_test_actual, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_actual, predictions)
r2 = r2_score(y_test_actual, predictions)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'R-squared (R²): {r2}')
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_test_actual, label='Actual Prices', color='blue')
plt.plot(predictions, label='Predicted Prices', color='red', linestyle='--')
plt.title('Actual vs Predicted Germany Prices')
plt.xlabel('Time steps')
plt.ylabel('Price')
plt.legend()
plt.show()
```
